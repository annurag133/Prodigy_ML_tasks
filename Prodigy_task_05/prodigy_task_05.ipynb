```python
data=pd.read_csv('nutrients_csvfile.csv')
```

```python
# Check for missing values
print(data.isnull().sum())

# Fill missing values with the mean of each column
data.fillna(data.mean(), inplace=True)

# Verify that there are no more missing values
print(data.isnull().sum())
```

```python
from sklearn.preprocessing import OneHotEncoder

# One-hot encode the 'Food' column
onehot_encoder = OneHotEncoder(sparse_output=False)
food_encoded = onehot_encoder.fit_transform(data[['Food']])

# Combine the encoded food names with the rest of the dataset
X = pd.DataFrame(food_encoded, columns=onehot_encoder.get_feature_names_out(['Food']))
y = data[['Calories', 'Protein', 'Fat', 'Sat.Fat', 'Fiber', 'Carbs']]

print(X.head())
print(y.head())
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```
